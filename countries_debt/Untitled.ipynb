```python
import pandas as pd
df = pd.read_csv('/home/hp/Downloads/data_sets/countries_debt/API_DT.AMT.DLXF.CD_DS2_en_csv_v2.csv').drop([0]).set_index('Data Source')
new_columns_names = list(df.iloc[0].values)
df.set_axis(new_columns_names, axis = 'columns', inplace=True)
df.drop(['Country Name'], inplace = True)
from matplotlib import pyplot as plt
#plt.bar(list(df.index)[:10], list(df[2008.0])[:10])
#plt.xticks(rotation=90)
#plt.show()
pak = df.loc['Pakistan']
debt = list(pak.values)[3:]
years = list(pak.keys()[3:])
for i in debt:
    if str(i) == 'nan':
        debt[debt.index(i)] = 0
am = ['Ayyub khan', 'Yehya khan', 'bhuttu', 'Fazal chodhri', 'Zia-ul-haq', 'Gulam ishaq', 'Laghari', 'Rafiq Tarar','Musharraf', 'Zardari', 'Nawaz']
colors = ['y']*10+['r']*2+['g']*2+['y']*5+['b']*10+['brown']*5+['r']*5+['g']*3+['y']*7+['r']*5+['b']*4
plt.figure(figsize=(15,6))
bar = plt.bar(years, debt,color=colors)
plt.title('Pakistan international Debt History', size = 25, color= 'r')
plt.ylabel('1000 Million $', size = 25, color='r')
plt.xticks(years, rotation=90)
height = 5848074000+300000000
color_2 = ['w', 'r', 'g', 'y', 'b', 'brown', 'r', 'g', 'y', 'r', 'b']
for i in range(10):
    plt.text(1960, height-300000000 , am[i], color=color_2[i], size = 12 )
    height -= 300000000
for i in bar: 
    plt.text(i.get_x(), i.get_height()+700000000, str(int(round(debt[bar.index(i)]/1000000)))+'M $', rotation = 90)
plt.text(1965, 5848074000, 'Data Sourse: World bank website [https://data.worldbank.org/indicator/DT.AMT.DLXF.CD]', color='b', size = 12)
plt.savefig('pakdebt.svg')
plt.show()


a = ['r', 'b', 'g', 'y', 'orange', 'purple']
aa = {'70-71': 'yahya khan', '73': 'bhuttu', '78': 'Fazal Ilahai Chohdari', '1988':'zia', '1993': 'Ghulam Ishaq Khan', '1998': 'Farooq Ahmed Khan Laghari', '2001': 'Rafiq Tarar', '2008':'musharraf', '2013': 'zardari'}
    
```
